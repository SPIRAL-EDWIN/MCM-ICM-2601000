```python
import numpy as np
import matplotlib.pyplot as plt

# 参数
p = 10000
r = 1.0
x0, y0 = 1.0, 1.0

# 一次性生成所有点
px = np.random.rand(p) * 2.0
py = np.random.rand(p) * 2.0

# 是否在圆内
inside = (px - x0)**2 + (py - y0)**2 < r**2
n = np.count_nonzero(inside)

# 可视化
plt.figure()
plt.scatter(px[inside],  py[inside],  s=4, marker='.', label='inside',  c='b')
plt.scatter(px[~inside], py[~inside], s=4, marker='.', label='outside', c='r')

plt.gca().set_aspect('equal', adjustable='box')
plt.xlim(0, 2)
plt.ylim(0, 2)
plt.legend()
plt.title('Monte Carlo π estimation (vectorized)')
plt.show()

# 估计 π：单位正方形面积为 4，圆面积约为 πr^2 = π
pi0 = (n / p) * 4.0
print("pi0 =", pi0)
```
