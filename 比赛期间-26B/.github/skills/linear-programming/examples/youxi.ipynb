```python
import numpy as np
from scipy.optimize import linprog

# 目标函数系数 (带负号表示最大化问题转为最小化)
c = np.array([-20, -30, -45])

# 不等式约束 A_ub x <= b_ub
A_ub = np.array([
    [4, 8, 15],
    [1, 1, 1]
])
b_ub = np.array([100, 20])

# 变量下界（lb = 0）
bounds = [(0, None) for _ in range(3)]

# 调用 linprog
res = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method="highs")

if res.success:
    x = res.x               # 最优解
    fval = res.fun          # 线性规划的最小值（因为是 -max）
    y = -fval               # 目标函数最大化的结果

    print("A、B、C三图分别通关的次数为：")
    print(x)
    print("最终获得的经验为：")
    print(y)
else:
    print("未找到最优解")
```
